```python
count = 1
# Membaca semua gambar pada folder dengan fungsi glob
for filename in glob.glob('I:\KULIAH\Semester_8\PROPOSAL-TA\Dataset\Dataset Asli\Matang\Latihan\Crop\*.jpg'):
    myimg = cv2.imread(filename)
    img = cv2.cvtColor(myimg,cv2.COLOR_BGR2RGB)
    img = cv2.cvtColor(myimg,cv2.COLOR_RGB2YCR_CB)
    avg_color_per_row = numpy.average(img, axis=0)
    avg_color = numpy.average(avg_color_per_row, axis=0)
    # color format is BGR not RGB, cv2 membaca gambar dalam format BGR.
    # save into 1 matrix
    if count == 1 :
        arr = np.hstack([avg_color,['Matang']])
        count = 2
    else :
        row = np.array(avg_color)
        row = np.hstack([row,['Matang']])
        arr = np.vstack([arr,row])
print("Hasil Data 1 : ")
print(arr)

```
